Calculate the Power and Energy in Python

```python
import numpy as np
import scipy as scipy
```

Energy Formula



```python
# Formula adapted from matlab: https://stackoverflow.com/a/34133755/5504438
# Need to verify that it is correct
def get_energy(x):
    Xf = scipy.fft(x)
    return np.sum(np.abs(Xf) ** 2) / Xf.size
```

```python
import pandas as pd
testData = pd.read_csv('data1.csv')
inputData = testData['gyroX'][:3]
print(inputData)
get_energy(inputData)
```
